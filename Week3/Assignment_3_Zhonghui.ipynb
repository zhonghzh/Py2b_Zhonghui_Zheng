```python
import numpy as np
import pandas as pd
df = pd.read_csv("winequality-red.csv")
df
```

```python
samples=len(df)
columns=len(df.columns)
print(samples,columns)
```

```python
def count_nan():
        for col in df.columns:
            column_name = col
            working_col = df[col]
            working_col_missing_samp = working_col.isna().sum()
            print(f"The column {column_name} is missing {working_col_missing_samp/columns*100}% of data! ")        
count_nan()
```

```python
nan_samples = df[df.isna().all(axis=1)]
nan_indexes = nan_samples.index.values
df = df.drop(nan_indexes)
count_nan()
```

```python
df = df.sort_values("quality")
df = df.reset_index(drop=True)
df
```

```python
def normalize_values(col_value, min_val, max_val):
    normalized_val = np.divide(col_value - min_val, max_val - min_val)
    return normalized_val

def main():
    for column in df.columns:
        col_value = df[column]
        min_val = df[column].min()
        max_val = df[column].max()
        df[column] = normalize_values(col_value, min_val, max_val)


main()
df.describe()
```
